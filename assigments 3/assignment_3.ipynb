```python
#exercise_lgis
def longest_increasing_subsequence(perm):
    n = len(perm)                                                            #number of elements in the permutation
    lis_lengths = [1] * n  
    prev_index = [-1] * n  
    
    for i in range(n):
        for j in range(i):
            if perm[i] > perm[j] and lis_lengths[i] < lis_lengths[j] + 1:
                lis_lengths[i] = lis_lengths[j] + 1                          #update lis_lengths
                prev_index[i] = j                                            #set the previous index in the lis

    max_len = max(lis_lengths)                                               #find the maximum length of the longest increasing subsequence from the lis_lengths array.
    idx = lis_lengths.index(max_len)                                         #here we get index of the element with the maximum length in lis_lengths

    lis = []
    while idx != -1:
        lis.append(perm[idx])                                                #append the current element to the LIS
        idx = prev_index[idx]                                                #move to the previous element in the LIS

    lis.reverse()                                                            #reverse the sequence since we built it backwards
    return lis

def longest_decreasing_subsequence(perm):
    n = len(perm)
    lds_lengths = [1] * n  
    prev_index = [-1] * n  
    
    for i in range(n):
        for j in range(i):
            if perm[i] < perm[j] and lds_lengths[i] < lds_lengths[j] + 1:
                lds_lengths[i] = lds_lengths[j] + 1                         #update the length of LDS ending at perm[i]
                prev_index[i] = j                                           #we set the previous element index in the LDS

    max_len = max(lds_lengths)
    idx = lds_lengths.index(max_len)

    lds = []
    while idx != -1:
        lds.append(perm[idx]) 
        idx = prev_index[idx]

    lds.reverse()                                                          #we have to reverse the sequence since we built it backwards
    return lds

with open('rosalind_lgis.txt', 'r') as file:
    n = int(file.readline()) 
    perm = list(map(int, file.readline().split()))

lis = longest_increasing_subsequence(perm)
lds = longest_decreasing_subsequence(perm)

print(" ".join(map(str, lis)))
print(" ".join(map(str, lds)))
```

```python
#exercise_edita
def read_fasta(file_path):
    #reads a FASTA file and returns a list of sequences
    with open(file_path, 'r') as file:
        lines = file.readlines()

    sequences = []
    current_sequence = []
    for line in lines:
        line = line.strip()                                     #remove any whitespace or newline characters
        if line.startswith(">"):
            if current_sequence:
                sequences.append(''.join(current_sequence))
            current_sequence = []
        else:
            current_sequence.append(line)
    
    if current_sequence:
        sequences.append(''.join(current_sequence))

    return sequences

def edit_distance_alignment(s, t):
    #calculates the edit distance and optimal alignment between two strings s and t
    m, n = len(s), len(t)
    
    # Initialize the DP table with zeros
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    
    #fill base cases for deletions and insertions
    for i in range(m + 1):
        dp[i][0] = i
    for j in range(n + 1):
        dp[0][j] = j
    
    # fill the DP table using the recurrence relations
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s[i - 1] == t[j - 1]:
                dp[i][j] = dp[i - 1][j - 1]
            else:
                dp[i][j] = min(
                    dp[i - 1][j - 1], 
                    dp[i][j - 1],      
                    dp[i - 1][j]       
                ) + 1
    
    edit_distance = dp[m][n]
    
    #trace back to get the optimal alignment
    s_prime, t_prime = [], []
    i, j = m, n
    while i > 0 or j > 0:
        if i > 0 and j > 0 and s[i - 1] == t[j - 1]:
            s_prime.append(s[i - 1])
            t_prime.append(t[j - 1])
            i -= 1
            j -= 1
        elif i > 0 and j > 0 and dp[i][j] == dp[i - 1][j - 1] + 1:
            s_prime.append(s[i - 1])
            t_prime.append(t[j - 1])
            i -= 1
            j -= 1
        elif i > 0 and dp[i][j] == dp[i - 1][j] + 1:
            #deletion from s
            s_prime.append(s[i - 1])
            t_prime.append('-')
            i -= 1
        else:
            #insertion into s
            s_prime.append('-')
            t_prime.append(t[j - 1])
            j -= 1
    
    #we have to reverse the lists since we have built them backwards
    s_prime.reverse()
    t_prime.reverse()

    return edit_distance, ''.join(s_prime), ''.join(t_prime)

#main function to process the input and compute the edit distance and alignment
def main():
    file_path = "rosalind_edta.fasta"  
    sequences = read_fasta(file_path)
    
    if len(sequences) == 2:
        s = sequences[0]  
        t = sequences[1] 
        
        distance, s_prime, t_prime = edit_distance_alignment(s, t)
        
        print(distance)
        print(s_prime)
        print(t_prime)
    else:
        print("The FASTA file must contain exactly two sequences") 
main()
```

```python
# exercise_ctea
MOD = 134217727  #used to prevent overflow and keep the results within a manageable range

def read_fasta(file_path):                             #read the fasta file and return a list of sequences
    sequences = []
    with open(file_path, 'r') as f:                    #open the file for reading
        sequence_data = []
        for line in f:                                 #iterate through each line in the file
            line = line.strip()                        #remove leading/trailing whitespace
            if line.startswith(">"):  
                if sequence_data:  
                    sequences.append(''.join(sequence_data))
                sequence_data = [] 
            else:
                sequence_data.append(line)             #collect sequence data
        if sequence_data: 
            sequences.append(''.join(sequence_data))   #add the last sequence to the list
    return sequences

def number_of_optimal_alignments(s, t):                      #function to calculate the number of optimal alignments between two strings s and t
    m, n = len(s), len(t)
    
    dp = [[float('inf')] * (n + 1) for _ in range(m + 1)]    #array for storing minimum edit distance
    
    count = [[0] * (n + 1) for _ in range(m + 1)]            #array for storing the count of optimal alignments
    
    #base case: empty string alignments
    dp[0][0] = 0
    count[0][0] = 1
    
    #fill DP table
    for i in range(m + 1):
        for j in range(n + 1):
            if i > 0 and j > 0:
                match_cost = 0 if s[i - 1] == t[j - 1] else 1  #0 for match, 1 for mismatch
                new_cost = dp[i - 1][j - 1] + match_cost
                if new_cost < dp[i][j]:
                    dp[i][j] = new_cost
                    count[i][j] = count[i - 1][j - 1]
                elif new_cost == dp[i][j]:
                    count[i][j] = (count[i][j] + count[i - 1][j - 1]) % MOD
            
            if i > 0:
                new_cost = dp[i - 1][j] + 1                                #deletion cost from the cell above
                if new_cost < dp[i][j]:
                    dp[i][j] = new_cost
                    count[i][j] = count[i - 1][j]
                elif new_cost == dp[i][j]:
                    count[i][j] = (count[i][j] + count[i - 1][j]) % MOD
            
            if j > 0:
                new_cost = dp[i][j - 1] + 1                                #insertion cost from the cell to the left
                if new_cost < dp[i][j]:
                    dp[i][j] = new_cost
                    count[i][j] = count[i][j - 1]
                elif new_cost == dp[i][j]:
                    count[i][j] = (count[i][j] + count[i][j - 1]) % MOD
    
    return count[m][n]

file_path = 'rosalind_ctea.fasta'
sequences = read_fasta(file_path)

#check that we have at least two sequences
if len(sequences) < 2:
    print("Error: The FASTA file must contain at least two sequences.")
else:
    #use the first two sequences from the file
    s = sequences[0]
    t = sequences[1]

    result = number_of_optimal_alignments(s, t)
    print(result)

```

```python
# BIOPYTHON: 
# Add your functions here
from Bio import pairwise2
from Bio import SeqIO

def read_sequence(file_path):
    record = SeqIO.read(file_path, "fasta")
    return record.seq


#FUNCTION 1
def countMatches(s1, s2):
    return sum(1 for x, y in zip(s1, s2) if x == y and x != '-' and y != '-')

#FUNCTION 2
def countMismatches(s1, s2):
    return sum(1 for x, y in zip(s1, s2) if x != y and x != '-' and y != '-')

#FUNCTION 3
def countGapOpens(s1, s2):
    count = 0
    #loop through the sequences to detect where gaps start
    for i in range(1, len(s1)):
        if s1[i] == '-' and s1[i-1] != '-':
            count += 1
        if s2[i] == '-' and s2[i-1] != '-':
            count += 1
    #if there's a gap at the first position, count it as a gap open
    if s1[0] == '-' or s2[0] == '-':
        count += 1
    return count

#FUNCTION 4
#count the number of gap extensions in two aligned sequences
def countGapExtensions(s1, s2):
    count = 0
    #loop through the sequences to detect consecutive gaps
    for i in range(1, len(s1)):
        if s1[i] == '-' and s1[i-1] == '-':
            count += 1
        if s2[i] == '-' and s2[i-1] == '-':
            count += 1
    return count

#FUNCTION 5
#calculating the alignment score based on matches, mismatches, gap openings, and gap extensions
def getScore(s1, s2, matchScore, mismatchPenalty, gapOpenPenalty, gapExtensionPenalty):
    score = 0
    gap_opened = False 
    for i in range(len(s1)):
        if s1[i] == '-' or s2[i] == '-':
            if not gap_opened:
                score += gapOpenPenalty  
                gap_opened = True
            else:
                score += gapExtensionPenalty  
        else:
            if s1[i] == s2[i]:
                score += matchScore 
            else:
                score += mismatchPenalty  
            gap_opened = False 
    return score

#function to align the sequences
def align_sequences(s1, s2, matchScore, mismatchPenalty, gapOpenPenalty, gapExtensionsPenalty):  
    from Bio import pairwise2
    alignments = pairwise2.align.globalms(s1, s2, matchScore, mismatchPenalty, gapOpenPenalty, gapExtensionsPenalty)
    return alignments  

#function to extract aligned sequences and score
def getScores(alignments):
    aligned_s1, aligned_s2 = alignments[0][:2]
    alignment_score = alignments[0][2]
    return aligned_s1, aligned_s2, alignment_score

#sequences from FASTA file
sequence_a = read_sequence("IL12A.fasta")
sequence_b = read_sequence("IL12B.fasta")


#parametes for the score 
match_score = 1
mismatch_penalty = -1
gap_open_penalty = -0.5
gap_extension_penalty = -0.2

#perform the alignment
alignments = align_sequences(sequence_a, sequence_b, match_score, mismatch_penalty, gap_open_penalty, gap_extension_penalty)

#extract the aligned sequences and the score
aligned_s1, aligned_s2, score = getScores(alignments)

print("Risultati dalla funzione getScores:")
print(f"Punteggio dell'allineamento (da getScores): {score}")

#calculate the various parameters using the defined functions
matches = countMatches(aligned_s1, aligned_s2)
mismatches = countMismatches(aligned_s1, aligned_s2)
gap_opens = countGapOpens(aligned_s1, aligned_s2)
gap_extensions = countGapExtensions(aligned_s1, aligned_s2)

print("\nAlignment analysis results:")
print(f"Number of matches: {matches}")
print(f"Number of mismatches: {mismatches}")
print(f"Number of gap openings: {gap_opens}")
print(f"Number of gap extensions: {gap_extensions}")

#calculate the score manually using the getScore function
manual_score = getScore(aligned_s1, aligned_s2, match_score, mismatch_penalty, gap_open_penalty, gap_extension_penalty)
print(f"alignment score (from getScore): {manual_score}")

#direct alignment using pairwise2
alignments_pairwise = pairwise2.align.globalms(sequence_a, sequence_b, match_score, mismatch_penalty, gap_open_penalty, gap_extension_penalty)
aligned_s1_pairwise, aligned_s2_pairwise = alignments_pairwise[0][:2]
score_pairwise = alignments_pairwise[0][2]

print("\nRisultati da pairwise2:")
print(f"alignment score (from pairwise2): {score_pairwise}")
```
